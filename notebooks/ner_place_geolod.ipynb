```python
def extract_place_and_geolocation(text: str):

    results = api.geoparse(text)

    list_latitude = []
    list_longitude = []
    list_places_title = []


    # find the entities that annoate as place from geoLOD
    for entity in results:
        if entity['geometry']!=None:
            place = entity['properties']['surface']
            longitude = entity['properties']['geoword_properties']['longitude']
            latitude = entity['properties']['geoword_properties']['latitude']

            #print(place)
            #print('latitude:', latitude)
            #print('longitude:', longitude)
            
            list_longitude.append(longitude)
            list_latitude.append(latitude)
            list_places_title.append(place)

    df = pd.DataFrame({'Location':list_places_title,
                       'Latitude':list_latitude,
                       'Longitude':list_longitude})

    return df
```

```python
def print_meta_info_for_excluded_places(title: str, list_places_title: list, model_name: str = 'GeoLOD'):

    # find the exluded place 
    print('The excluded places that appear in the title:')
    for exclude_place in excluded_location.values():
        if exclude_place in title:
            if exclude_place not in list_places_title: # find the exluded place that not appear in title
                value = {i for i in excluded_location if excluded_location[i]==exclude_place} # key by value:
                print(f'Correct not annotate (from {model_name})' , colored(str(value) + ' ' + exclude_place, 'green') )
            else:
                value = {i for i in excluded_location if excluded_location[i]==exclude_place} # key by value:
                print(f'Not Correct annotate (from {model_name})' , colored(str(value) + ' ' + exclude_place, 'red') )

    return
```
